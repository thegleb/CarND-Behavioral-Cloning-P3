# Some visualization code to check out the models

Visualization code heavily based on https://github.com/gabrielpierobon/cnnshapes/blob/master/README.md

It has been heavily modularized and tweaked slightly to fit the irregularly shaped outputs (rectangular instead of square).

```python
import matplotlib.pyplot as plt
import numpy as np
from keras import Sequential
from keras.layers import Lambda, Cropping2D, Conv2D, Dropout, Dense, Flatten
from keras.layers.pooling import MaxPooling2D
from keras.models import Model
from keras.preprocessing import image

def get_image_tensor(img_path):
    """
    Given an image path, generate a Keras-ready image tensor
    """
    img = image.load_img(img_path)
    img_tensor = image.img_to_array(img)
    img_tensor = np.expand_dims(img_tensor, axis=0)
    # optional step, can cut down on noise in the visualization, but also lose a bit of potentially valuable data
    # img_tensor /= 255.
    return img_tensor
    
def draw_activations(layer_names, activations, prefix, model_name, images_per_row=4):
    """
    Draw activations for each layer
    """
    for layer_name, layer_activation in zip(layer_names, activations):  # Displays the feature maps
        n_features = layer_activation.shape[-1]  # Number of features in the feature map
        shape = (layer_activation.shape[1], layer_activation.shape[2])  # The feature map has shape (1, size, size, n_features).
        n_cols = n_features // images_per_row  # Tiles the activation channels in this matrix
        display_grid = np.zeros((n_cols * shape[0], images_per_row * shape[1]))
        for col in range(n_cols):  # Tiles each filter into a big horizontal grid
            for row in range(images_per_row):
                channel_image = layer_activation[0, :, :, col * images_per_row + row]
                channel_image -= channel_image.mean()  # Post-processes the feature to make it visually palatable
                channel_image /= channel_image.std()
                channel_image *= 255
                channel_image = np.clip(channel_image, 0, 255).astype('uint8')
                display_grid[
                    col * shape[0]: (col + 1) * shape[0],
                    row * shape[1]: (row + 1) * shape[1]
                ] = channel_image
        scale = 1. / shape[0]
        plt.figure(figsize=(scale * display_grid.shape[1],
                            scale * display_grid.shape[0]))
        plt.title(layer_name)
        plt.grid(False)
        plt.imsave('{}-{}-{}.png'.format(prefix, model_name, layer_name), display_grid, cmap='gray')

def visualize_model(model, img_path, model_name):
    """
    Given a model and an image path, draw the layer activations for the input image
    """
    # these layers of interest are common between the models
    layers = model.layers[slice(2, 6)]
    layer_outputs = [layer.output for layer in layers]
    layer_names = [layer.name for layer in layers]
    img_tensor = get_image_tensor(img_path)

    activation_model = Model(inputs=model.input, outputs=layer_outputs)
    activations = activation_model.predict(img_tensor)

    draw_activations(layer_names, activations, img_path.rsplit('.', 1)[0], model_name)
```

```python
# new model, minus dropout layers to make visualization a bit easier
model = Sequential([
    Lambda(lambda x: x / 255.0 - 0.5, input_shape=(160, 320, 3), output_shape=(160, 320, 3)),
    Cropping2D(cropping=((60, 25), (0, 0))),
    Conv2D(filters=24, kernel_size=(5, 5), activation='relu', strides=(2, 2)),
    Conv2D(filters=36, kernel_size=(5, 5), activation='relu', strides=(2, 2)),
    Conv2D(filters=48, kernel_size=(5, 5), activation='relu', strides=(2, 2)),
    Conv2D(filters=64, kernel_size=(3, 3), activation='relu'),
    Conv2D(filters=64, kernel_size=(3, 3), activation='relu'),
    Flatten(),
    Dense(100),
    Dense(50),
    Dense(10),
    Dense(1)
])
model.load_weights('model.h5')

# original lenet model
model_lenet = Sequential([
    Lambda(lambda x: x / 255.0 - 0.5, input_shape=(160, 320, 3)),
    Cropping2D(cropping=((55,25), (0,0))),
    Conv2D(filters=6, kernel_size=(5, 5), activation='relu'),
    MaxPooling2D(),
    Conv2D(filters=16, kernel_size=(5, 5), activation='relu'),
    MaxPooling2D(),
    Flatten(),
    Dense(120),
    Dropout(0.3),
    Dense(84),
    Dense(1)
])
model_lenet.load_weights('model-track1-only.h5')
    
images = [
    './examples/track1-bridge.jpg',
    './examples/track1-curve.jpg',
    './examples/track1-middle-road.jpg',
    './examples/track1-reverse-near-dirt.jpg',
    './examples/track2-middle-road.jpg',
    './examples/track2-mountain.jpg',
    './examples/track2-offroad.jpg',
    './examples/track2-right-of-middle.jpg',
]
```

```python
for i in range(len(images)):
    img_path = images[i]
    visualize_model(model_lenet, img_path, 'lenet')
```
